```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

## Iris

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
x = iris.drop('species', axis=1).values
y = iris['species'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
kf = KFold(n_splits=3)
```

```python
acs1nw =[]
acs10wd =[]
acs10wu =[]
```

```python
for train_index, val_index in kf.split(x_train):
    x_train_kf, x_val = x_train[train_index], x_train[val_index]
    y_train_kf, y_val = y_train[train_index], y_train[val_index]

    #para knn = 1 sin pesos
    print('knn = 1 sin pesos')
    knn1 = KNeighborsClassifier(n_neighbors=1)
    knn1.fit(x_train_kf, y_train_kf)
    y_pred = knn1.predict(x_val)
    ac1nw = accuracy_score(y_val, y_pred)
    print(ac1nw)
    acs1nw.append(ac1nw)

    #para knn = 10 con pesos por distancia
    print('knn = 10 con pesos por distancia')
    knn10d = KNeighborsClassifier(n_neighbors=10, weights='distance')
    knn10d.fit(x_train_kf, y_train_kf)
    y_pred = knn10d.predict(x_val)
    ac10wd = accuracy_score(y_val, y_pred)
    print(ac10wd)
    acs10wd.append(ac10wd)

    #para knn = 10 con peso uniforme
    print('knn = 10 con peso uniforme')
    knn10u = KNeighborsClassifier(n_neighbors=10, weights='uniform')
    knn10u.fit(x_train_kf, y_train_kf)
    y_pred = knn10u.predict(x_val)
    ac10wu = accuracy_score(y_val, y_pred)
    print(ac10wu)
    acs10wu.append(ac10wu)

average_acs1nw = np.mean(acs1nw)
acs1nw.append(average_acs1nw)

average_acs10wd = np.mean(acs10wd)
acs10wd.append(average_acs10wd)

average_acs10wu = np.mean(acs10wu)
acs10wu.append(average_acs10wu)

print('Promedio de accuracy para knn = 1 sin pesos:', average_acs1nw)
print('Promedio de accuracy para knn = 10 con pesos por distancia:', average_acs10wd)
print('Promedio de accuracy para knn = 10 con pesos uniformes:', average_acs10wu)

if average_acs1nw > average_acs10wd and average_acs1nw > average_acs10wu:
    best_model = knn1
    print('El mejor modelo es knn = 1 sin pesos')
elif average_acs10wd > average_acs1nw and average_acs10wd > average_acs10wu:
    print('El mejor modelo es knn = 10 con pesos por distancia')
    best_model = knn10d
else:
    print('El mejor modelo es knn = 10 con pesos uniformes')
    best_model = knn10u

best_model.fit(x_train, y_train)
y_pred = best_model.predict(x_test)
ac = accuracy_score(y_test, y_pred)
print('Accuracy en test:', ac)

report = classification_report(y_test, y_pred)
print('Reporte de clasificación:')
print(report)

cm = confusion_matrix(y_test, y_pred)
cmd = ConfusionMatrixDisplay(cm, display_labels=best_model.classes_)
cmd.plot()
plt.show()
```
